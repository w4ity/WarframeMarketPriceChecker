```python
import statistics
import requests
import pandas as pd
from concurrent.futures import ThreadPoolExecutor, as_completed

BASE_URL = "https://api.warframe.market"
HEADERS = {
    "Accept": "application/json",
    "Language": "en",
    "Platform": "pc",
}
session = requests.Session()
session.headers.update(HEADERS)

def analizar_item(slug):
    try:
        url = f"{BASE_URL}/v2/orders/item/{slug}"
        r = session.get(url, timeout=5)
        r.raise_for_status()
        
        data = r.json()["data"]
        online = [
            o for o in data
            if o["user"]["status"] in ("online", "ingame")
        ]
        sells = sorted(
            o["platinum"] for o in online
            if o["type"] == "sell"
        )
        top_sells = sells[:5]  
        return {
            "item": slug,
            "min_sell": sells[0] if sells else 0,
            "avg_sell": round(statistics.median(top_sells), 2) if top_sells else 0,
            "sells_used_for_avg": top_sells 
        }
    except Exception as e:
        return {
            "item": slug,
            "min_sell": 0,
            "avg_sell": 0,
            "sells_used_for_avg": [],
            "error": str(e)
        }

articulos = [
    "galvanized_acceleration", 
    "galvanized_savvy", 
    "galvanized_chamber",
    "galvanized_aptitude", 
    "galvanized_crosshairs", 
    "galvanized_diffusion",
    "galvanized_elementalist", 
    "galvanized_scope", 
    "galvanized_hell",
    "galvanized_shot", 
    "galvanized_reflex", 
    "galvanized_steel",
    'aerodynamic',
    'energizing_shot',
    'mending_shot', 
    'aerial_ace', 
    'preparation',
    'rolling_guard',
    'vigorous_swap',
    'sharpshooter',
    'cautious_shot',
    'power_donation'
]

with ThreadPoolExecutor(max_workers=15) as executor:
    futures = {executor.submit(analizar_item, a): a for a in articulos}
    resultados = [future.result() for future in as_completed(futures)]

df = pd.DataFrame(resultados)
df = df.sort_values('avg_sell', ascending=False).reset_index(drop=True)
df["Copy"] = df["item"].str.replace("_", " ").str.title() 

pd.set_option('display.expand_frame_repr', False)
pd.set_option('display.max_colwidth', None)

display(df)
```
